# 2. Handling Missing Values of "Categorical features":

```python
# 2. Handling Missing Values of "Categorical features"


#out of all the features of dataset extracting missing values only for categorical features
feature_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=='O']

for i in feature_nan:
    print(i,"{}%".format(np.round(df[i].isnull().mean(),4)))
```

## 2.1 Replacing Missing values with a new label:

```python
# 2.1 Replacing Missing values with a new label

def replace_cat(df,feature_nan):
    data=df.copy()
    
    #replacing NAN with Missing coz we're trying to convert those nan values into a new category
    data[feature_nan]=df[feature_nan].fillna('Missing')
    return data

df=replace_cat(df,feature_nan)

df[feature_nan].isnull().sum()
```

# 3. Handling missing values for Numerical features:

```python
# 3. Handling missing values for Numerical features:

num_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']

for i in num_nan:
    print(i,"{}%".format(np.around(df[i].isnull().mean(),4))) #.around return array of round up values 
```

## <font color='purple'> In EDA we see that there are lots of outliers in num_features. So incase when you have lots of outliers in your features then you're supposed to replace the nan with `median` or `mode`. </font>

## 3.1 Replacing missing values with median:

```python
for i in num_nan:
    median_value=df[i].median()
    
    #Creating a new feature [feature+nan] and in that we're storing all the nan values of all the features in the datset in such a way by making nan value as 1 else 0
    #It is a good practice to create 3 columns separately representing only presence of missing values
    df[i+'nan']=np.where(df[i].isnull(),1,0) 
    df[i].fillna(median_value,inplace=True)

df[num_nan].isnull().sum()
```

# 4. Handling Temporal Variable:


We've seen the relationship of year_sold w.r.t SalePrice in EDA, which says that, As the year_sold increases the SalePrice was decreasing.

```python
# 5. To Handle Temporal Variable 
# Converting the values in columns ('YearBuilt','YearRemodAdd','GarageYrBlt') to no. of years

for i in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    df[i]=df['YrSold']-df[i]
```

# 5. Feature Transformation:
Since the numerical variables are skewed we'll perform Log Norm. Dist.

```python
#From EDA it is evident that below listed features have skewed value and does not contains any 0 values
#If any feature has 0 value then you're supposed to just skip it and not to perform Log Norm. Dist. on that feature

num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for i in num_features:
    df[i]=np.log(df[i])
```

# 6. Handling Rare Categorical Features:
Remove those Feature variables with less than 1% of occurance 

```python
categorical_features=[feature for feature in df.columns if df[feature].dtypes=='O']

categorical_features
```

1. **temp=df.groupby(i)['SalePrice'].count()/len(df):**
    * df.groupby(i)['SalePrice']: Grouping by each variable inside the feature w.r.t SalePrice
    * df.groupby(i)['SalePrice'].count(): After Grouping counting
    * df.groupby(i)['SalePrice'].count()/len(df): Divide it with total length of entire dataset
    * temp=df.groupby(i)['SalePrice'].count()/len(df): And store it in the variable temp

2. **temp_df=temp[temp>0.01].index:** 
    * Wherever the value of temp>0.01 then take it's index and store it in temp_df

3. **df[i]=np.where(df[i].isin(temp_df),df[i],'Rare_var'):**
    * If df[i] is in temp_df then take that variable (i) else make that value as 'Rare_var'

```python
for i in categorical_features:
    temp=df.groupby(i)['SalePrice'].count()/len(df) #Calculating the % of each variable of a feature
    temp_df=temp[temp>0.01].index
    df[i]=np.where(df[i].isin(temp_df),df[i],'Rare_var')
    
```

```python
df.head(40)
```

```python
for i in categorical_features:
    labels_ordered=df.groupby([i])['SalePrice'].mean().sort_values().index
    labels_ordered={k:x for x, k in enumerate(labels_ordered,0)}
    df[i]=df[i].map(labels_ordered)
```
